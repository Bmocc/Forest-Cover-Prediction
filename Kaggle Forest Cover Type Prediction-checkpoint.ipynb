## Data Interaction
    Correlation

```python
import numpy

# Correlation tells relation between two attributes.
# Correlation requires continous data. Hence, ignore Wilderness_Area and Soil_Type as they are binary

#sets the number of features considered
size = 10 

#create a dataframe with only 'size' features
data=train_df.iloc[:,:size] 

#get the names of all the columns
cols=data.columns 

# Calculates pearson co-efficient for all combinations
data_corr = data.corr()

# Set the threshold to select only only highly correlated attributes
threshold = 0.5

# List of pairs along with correlation above threshold
corr_list = []

#Search for the highly correlated pairs
for i in range(0,size): #for 'size' features
    for j in range(i+1,size): #avoid repetition
        if (data_corr.iloc[i,j] >= threshold and data_corr.iloc[i,j] < 1) or (data_corr.iloc[i,j] < 0 and data_corr.iloc[i,j] <= -threshold):
            corr_list.append([data_corr.iloc[i,j],i,j]) #store correlation and columns index

#Sort to show higher ones first            
s_corr_list = sorted(corr_list,key=lambda x: -abs(x[0]))

#Print correlations and column names
for v,i,j in s_corr_list:
    print ("%s and %s = %.2f" % (cols[i],cols[j],v))

# Strong correlation is observed between the following pairs
# This represents an opportunity to reduce the feature set through transformations such as PCA
```

## Data Interaction
    Scatter Plot

```python
#import plotting libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter plot of only the highly correlated pairs
for v,i,j in s_corr_list:
    sns.pairplot(train_df, hue="Cover_Type", height=6, x_vars=cols[i],y_vars=cols[j] )
    plt.show()

#The plots show to which class does a point belong to. The class distribution overlaps in the plots.    
#Hillshade patterns give a nice ellipsoid patterns with each other
#Aspect and Hillshades attributes form a sigmoid pattern
#Horizontal and vertical distance to hydrology give an almost linear pattern.
```

### Data Preperation <a class="anchor" id="DataPrep"></a>



#### preprocessing <a class="anchor" id="section_3_1"></a>

Preparing the data for modeling is important since it can speed up the process and produce better models. As the adage goes, "garbage in garbage out" so we want to make sure the data we are inputing into our modelling step is good enough to share with others.

First I will drop the columns with standard deviation of 0. Removing unnecessary columns reduces complexity and adheres to the Occam's Razor principle. 

```python
# columns with std = 0    
rm = train_df.std()[train_df.std() == 0].index.values

# drop the columns 
train_df.drop(rm, axis=1,inplace=True)
```

Split data in to train and test 

```python
col_length = len(train_df.columns)

#Y is the target column, X has the rest
X = train_df.iloc[:, 0:col_length-1]
Y = train_df.iloc[:, col_length-1:col_length]

#Validation chunk size
val_size = 0.2

#Use a common seed in all experiments so that same chunk is used for validation
seed = 0

#Split the data into chunks
from sklearn.model_selection import train_test_split 
X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size=val_size, random_state=seed)
```

Categorical variables are already taken care of in the data so I don't have to encode them. 

Normalizing the dataset with the `StandardScaler`. It's often a good idea to normalize your data and a common way is through standardization. Here the mean will be 0 and a standard deviation of 1 will be used on the columns. 

```python
#Import libraries for data transformations
from sklearn.preprocessing import StandardScaler

#All features
X_all = []

#Add this version of X to the list 
X_all.append(['Orig','All', X_train,X_val,1.0,cols[:col_length-1]])

#point where categorical data begins
size=10
```

Here we will run through the standard scaling and save the outputs in to `X_all`. It's important to only run the scaler for non-categorical columns. For example, data like elevation and other numeric data should be scaled. 

```python
#Standardized
#Apply transform only for non-categorical data
X_temp = StandardScaler().fit_transform(X_train.iloc[:,0:size])
X_val_temp = StandardScaler().fit_transform(X_val.iloc[:,0:size])
#Concatenate non-categorical data and categorical
X_con = np.concatenate((X_temp,X_train.iloc[:,size:]),axis=1)
X_val_con = np.concatenate((X_val_temp,X_val.iloc[:,size:]),axis=1)
#Add this version of X to the list 
X_all.append(['StdSca','All', X_con,X_val_con,1.0,cols[:col_length-1]])
```


### Prediction <a class="anchor" id="Prediction"></a>


#### model building <a class="anchor" id="section_4_1"></a>


Now I'm ready to create some models. Here is a list of [Multi class models](https://scikit-learn.org/stable/modules/multiclass.html) available in scikit learn. I'm not going to use all of them, but just four of them for now. 

I'm going to load all of the models and create a list called `models` to add the different ones I'll be using.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier

from sklearn import model_selection
# prepare models
models = []
models.append(('LR', LogisticRegression()))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier()))
```

Now I'm going to use `cross_val_score` from `model_selection` to test `accuracy` as a measure for each model. I only have two data sets to input (the original and standardized) through each model in this for loop. 

Below you'll see that `KNN` with the original data did the best with `CART` or decision tree with the original data in second place. 

```python
# evaluate each model in turn
results = []
names = []
scoring = 'accuracy'
seed = 5
for trans,name,X,X_val,v,cols_list in X_all:
    for modName, model in models:
        kfold = model_selection.KFold(n_splits=5, random_state=seed)
        cv_results = model_selection.cross_val_score(model, X, Y_train, cv=kfold, scoring=scoring)
        results.append((modName+trans,cv_results))
        names.append(modName+trans)
        msg = "%s: %f (%f)" % (modName+trans, cv_results.mean(), cv_results.std())
        print(msg)
```

##### hyperparameter tuning <a class="anchor" id="section_4_1_1"></a>

From the  default settings for the models, I wanted to see if I could get some minor improvements. I choose the `CART` algorithm since it had some hyperparameters I could change. Here is the description I collected about the `criterion` and `splitter` arguments. 

Tune decision tree

criterion : optional (default=â€giniâ€) or Choose attribute selection measure: This parameter allows us to use the different-different attribute selection measure. Supported criteria are â€œginiâ€ for the Gini index and â€œentropyâ€ for the information gain.

splitter : string, optional (default=â€bestâ€) or Split Strategy: This parameter allows us to choose the split strategy. Supported strategies are â€œbestâ€ to choose the best split and â€œrandomâ€ to choose the best random split.


```python
kfold = model_selection.KFold(n_splits=5, random_state=seed)
cv_results = model_selection.cross_val_score(DecisionTreeClassifier(criterion = 'entropy', splitter = 'random'),
                                             X_all[0][2], Y_train, cv=kfold, scoring='accuracy')
```

You can see that using `entropy` instead of `gini` and `random` instead of `best` in the `DecisionTreeClassifier` I was able to achieve modest gains. While it's difficult to say if this model will generalize well for the validation set in Kaggle, the cross validation does give hope.
